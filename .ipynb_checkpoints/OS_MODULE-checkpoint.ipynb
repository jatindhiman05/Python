**Renaming Folder**

```python
os.rename('note.txt','hello.txt')
```

**Printing all The info about hello.txt**

```python
os.stat('hello.txt')
```

```python
os.stat('hello.txt').st_size # size of file
```

```python
mod_time = os.stat('hello.txt').st_mtime # last modification time
print(datetime.fromtimestamp(mod_time)) # for human readable format
```

**Traversing Directory Tree**

```python
# useful for searching file in a folder on desktop but doesnot exactly remember the folder
for dirpath, dirnames, filenames in os.walk('/Users/Jatin Dhiman/Desktop/'):
    print(f'Current Path: ${dirpath}')
    print(f'Directories: ${dirnames}')
    print(f'files: ${filenames}')
    print()
```
